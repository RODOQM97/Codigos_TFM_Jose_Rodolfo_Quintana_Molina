## Random Forest - Indices de Vegetacion

```python
import pandas as pd

VI = pd.read_csv(r'C:\Users\Rodolfo\Documents\VIU\Tesis\Shapes\1.Datos_Indices\Arroto\Nuevo/DatasetALL_IV.csv')
VI
```

```python
VI['biomasa'].mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score,RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, make_scorer
import numpy as np

X = VI.drop(columns=['biomasa'])
y = VI['biomasa']
```

```python
X.head(5)
```

```python
y.head(5)
```

```python

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=42)

rf = RandomForestRegressor(random_state=42, n_jobs=-1)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)


cv_values = range(3, 6)
results_cv = []

for cv in cv_values:
    scores = cross_val_score(
        rf,
        X_train, y_train,
        cv=cv,
        scoring=mae_scorer,
        n_jobs=-1
    )
    mae_mean = -np.mean(scores)
    results_cv.append((cv, mae_mean))



best_cv, best_mae = min(results_cv, key=lambda x: x[1])
print("Mejor CV:", best_cv, "MAE promedio:", best_mae)

```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
import numpy as np

rf = RandomForestRegressor(random_state=42, n_jobs=-1)


param_dist = {
    'n_estimators': [400, 500, 600],       
    'max_depth': [4, 5, 6],                   
    'min_samples_split': [40, 50, 60],          
    'min_samples_leaf': [10, 20, 30],        
    'max_features': [0.15, 0.2, 0.25],         
    'criterion': ['squared_error']               
}

random_search_rf = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=60,                     
    cv=best_cv,                         
    scoring='neg_mean_absolute_error',  
    n_jobs=-1,
    random_state=42,
    verbose=1
)

random_search_rf.fit(X_train, y_train)

print("Mejores parámetros:", random_search_rf.best_params_)
print("Mejor puntuación (MAE):", -random_search_rf.best_score_)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


y_pred_train = random_search_rf.predict(X_train)
y_pred_val = random_search_rf.predict(X_val)

mae_train = mean_absolute_error(y_train, y_pred_train)
mae_val = mean_absolute_error(y_val, y_pred_val)

rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_val = np.sqrt(mean_squared_error(y_val, y_pred_val))

r2_train = r2_score(y_train, y_pred_train)
r2_val = r2_score(y_val, y_pred_val)

print("MAE Train:", mae_train, "MAE Val:", mae_val)
print("RMSE Train:", rmse_train, "RMSE Val:", rmse_val)
```

```python

rf_best_model_IV = RandomForestRegressor(
    n_estimators=500,
    min_samples_split=40,
    min_samples_leaf=10,
    max_features=0.25,
    max_depth=6,
    criterion='squared_error',
    random_state=42,
    n_jobs=-1
)

rf_best_model_IV.fit(X_train, y_train)
```

# SVR

## SVR - Indices de Vegetacion

```python
import pandas as pd

SVR_VI = pd.read_csv(r'C:\Users\Rodolfo\Documents\VIU\Tesis\Shapes\1.Datos_Indices\Arroto\Nuevo/DatasetALL_IV.csv')
SVR_VI
```

```python
SVR_VI['biomasa'].mean()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


X_SVR = SVR_VI.drop(columns=['biomasa'])
y_SVR = SVR_VI['biomasa']
```

```python
X_SVR.head(5)
```

```python
y_SVR.head(5)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_SVR, y_SVR, test_size=0.15, random_state=42)

pipeline_svr = Pipeline([
    ('scaler', StandardScaler()),
    ('model', SVR())   
])


mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)


cv_values = range(3, 6)
results_cv = []

for cv in cv_values:
    scores = cross_val_score(
        pipeline_svr,
        X_train, y_train,
        cv=cv,
        scoring=mae_scorer,
        n_jobs=-1
    )
    mae_mean = -np.mean(scores)
    results_cv.append((cv, mae_mean))



best_cv, best_mae = min(results_cv, key=lambda x: x[1])
print("Mejor CV:", best_cv, "MAE promedio:", best_mae)
```

```python

param_dist = {
    'model__C': [3.2, 3.4, 3.6],
    'model__gamma': [0.005, 0.01, 0.015],  # más pequeño
    'model__epsilon': [0.5, 1.0, 1.5],  # control de suavidad
    'model__kernel': ['rbf']
}

random_search_svr = RandomizedSearchCV(
    estimator=pipeline_svr,
    param_distributions=param_dist,
    n_iter=40,
    cv=best_cv,          
    scoring='neg_mean_absolute_error',        
    n_jobs=-1,
    random_state=42,
    verbose=1
)

random_search_svr.fit(X_train, y_train)

print("Mejores parámetros:", random_search_svr.best_params_)
print("Mejor puntacion MAE :", -random_search_svr.best_score_)
```

```python

y_pred_train = random_search_svr.predict(X_train)
y_pred_val = random_search_svr.predict(X_val)

mae_train = mean_absolute_error(y_train, y_pred_train)
mae_val = mean_absolute_error(y_val, y_pred_val)

rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_val = np.sqrt(mean_squared_error(y_val, y_pred_val))


print("MAE Train:", mae_train, "MAE Val:", mae_val)
print("RMSE Train:", rmse_train, "RMSE Val:", rmse_val)
```

```python
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

svr_best_model_IV = Pipeline([
    ('scaler', StandardScaler()),
    ('model', SVR(
        kernel='rbf',
        gamma=0.015,
        epsilon=0.5,
        C=3.6
    ))
])

svr_best_model_IV.fit(X_train, y_train)
```
