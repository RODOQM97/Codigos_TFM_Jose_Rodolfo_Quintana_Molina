```python

df = pd.read_csv(
    r'C:\Users\Rodolfo\Documents\VIU\Tesis\Shapes\1.Datos_Indices\Arroto\Datos_Arroto2020.csv'
)

df['date'] = pd.to_datetime(df['date'], dayfirst=True)

indices = [
    'GNDVI_S2', 'MSI_S2', 'NDRE_S2',
    'NDVI_S2', 'TCARIOSAVI_S2'
]

idx_sel = [1, 3, 6, 9, 12, 15, 18, 21, 24, 27, 30, 33, 35]

rows_final = []

for id_val, g in df.groupby('id'):
    g = g.sort_values('date').reset_index(drop=True)
    g['fecha_idx'] = g.index + 1
    g = g[g['fecha_idx'].isin(idx_sel)]

    row = {'id': id_val}

    for _, r in g.iterrows():
        i = int(r['fecha_idx'])
        row[f'date_{i}'] = r['date']
        for var in indices:
            base = var.replace('_S2', '')
            row[f'{base}_{i}'] = r[var]

    rows_final.append(row)

Arroto2020 = pd.DataFrame(rows_final)

```

```python
import numpy as np
import pandas as pd

biomasa_arroto = pd.read_csv(r'C:\Users\Rodolfo\Documents\VIU\Tesis\Shapes\1.Datos_Indices\Series_Tiempo\biomasa_arroto.csv')
biomasa_arroto.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = biomasa_arroto

df_long = df.melt(
    id_vars="id",
    value_vars=["Arroto_2020", "Arroto_2021", "Arroto_2022"],
    var_name="year",
    value_name="biomasa"
)

df_long["year"] = df_long["year"].str.replace("Arroto_", "").astype(int)


plt.figure(figsize=(10, 4))

sns.histplot(
    data=df_long,
    x="biomasa",
    hue="year",
    bins=45,

    palette=["tab:blue", "tab:orange", "tab:green"],
    alpha=0.4
)

plt.xlabel("Biomasa")
plt.ylabel("Frecuencia")
plt.title("Distribución de biomasa por año")

plt.tight_layout()



plt.show()
```

```python
df_long.groupby("year")["biomasa"].describe()
```

```python
resumen = df_long.groupby("year")["biomasa"].agg(
    mean="mean",
    std="std",
    min="min",
    max="max"
).reset_index()

resumen.to_csv(
    "C:/Users/Rodolfo/Documents/VIU/Tesis/Shapes/1.Datos_Indices/Series_Tiempo/metricas_arroto.csv",
    index=False
)
```

#### Grafica de correlacion de variables 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"C:\Users\Rodolfo\Documents\VIU\Tesis\Shapes\1.Datos_Indices\Arroto\Nuevo/DatasetALL_LST.csv")

corr = df.corr(numeric_only=True)

corr_sorted = corr.loc[corr.index, "biomasa"].sort_values(ascending=False).index
corr = corr.loc[corr_sorted, corr_sorted]

plt.figure(figsize=(18, 14))
sns.heatmap(
    corr,
    cmap="coolwarm",
    annot=False,
    linewidths=0.5,
    cbar_kws={"shrink": 0.7},
    center=0
)

plt.title("Matriz de correlación entre IV, LST y Biomasa", fontsize=18)
plt.tight_layout()

plt.savefig(
    r"C:\Users\Rodolfo\Documents\VIU\Tesis\Shapes\1.Datos_Indices\Figuras\correlaciones_variables.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv(r"C:\Users\Rodolfo\Documents\VIU\Tesis\Shapes\1.Datos_Indices\Arroto\Nuevo/DatasetALL_LST.csv")

corr = df.corr(numeric_only=True)['biomasa']

corr = corr.drop(labels=['biomasa'])

corr_strong = corr[(corr > 0.5) | (corr < -0.5)].sort_values()

plt.figure(figsize=(10, 8))
sns.barplot(x=corr_strong.values, y=corr_strong.index, palette="coolwarm")
plt.title("Variables fuertemente correlacionadas con la biomasa (|corr| > 0.5)", fontsize=16)
plt.xlabel("Correlación")
plt.ylabel("Variable")
plt.tight_layout()

plt.savefig(
    r"C:\Users\Rodolfo\Documents\VIU\Tesis\Shapes\1.Datos_Indices\Figuras\correlaciones_variables_mayores05.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```
